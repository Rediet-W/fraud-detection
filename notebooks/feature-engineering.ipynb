```python
import pandas as pd

# 1. Load the data
df_merged = pd.read_csv('../data/processed/cleaned_fraud_data.csv')

# 2. CRITICAL: Convert columns back to datetime objects
df_merged['signup_time'] = pd.to_datetime(df_merged['signup_time'])
df_merged['purchase_time'] = pd.to_datetime(df_merged['purchase_time'])

# 3. Now you can perform feature engineering without errors
df_merged['time_since_signup'] = (df_merged['purchase_time'] - df_merged['signup_time']).dt.total_seconds()
df_merged['hour_of_day'] = df_merged['purchase_time'].dt.hour
df_merged['day_of_week'] = df_merged['purchase_time'].dt.dayofweek

# 2. Velocity
# Count of transactions per device
device_counts = df_merged.groupby('device_id')['user_id'].transform('count')
df_merged['device_count'] = device_counts

# 3. Save Processed Data
df_merged.to_csv('../data/processed/merged_fraud_data.csv', index=False)
```
